```python
import requests
from bs4 import BeautifulSoup

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

url = "https://www.flipkart.com/apple-iphone-16-pro-max-desert-titanium-256-gb/p/itmb102b8a51100e?pid=MOBH4DQFTQHZAKAF&lid=LSTMOBH4DQFTQHZAKAFRC45AZ&marketplace=FLIPKART"

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

# Function to extract Product Title
def get_title(soup):
    try:
        title = soup.find("span", class_="VU-ZEz")  # Flipkart product title
        return title.text.strip()
    except:
        return ""

# Function to extract Product Price
def get_price(soup):
    try:
        price = soup.find("div", class_="Nx9bqj CxhGGd").text.strip()  # Flipkart price class
        return price
    except:
        return ""

# Function to extract Product Rating
def get_rating(soup):
    try:
        rating = soup.find("div", class_="XQDdHH").text.strip()  # Flipkart rating class
        return rating
    except:
        return ""

# Function to extract Number of User Reviews
def get_review_count(soup):
    try:
        review_count = soup.find("span", class_="Wphh3N").text.strip()  # Flipkart review count
        return review_count
    except:
        return ""

# Function to extract Availability Status
def get_availability(soup):
    try:
        availability = soup.find("div", class_="ZgI5Je").text.strip()  # Availability message
        return availability
    except:
        return "Not Available"

# --- Run extractors ---
print("Title:", get_title(soup))
print("Price:", get_price(soup))
print("Rating:", get_rating(soup))
print("Reviews:", get_review_count(soup))
print("Availability:", get_availability(soup))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# ---- Extractor Functions ----

def get_title(soup):
    try:
        title = soup.find("span", class_="VU-ZEz")
        return title.text.strip()
    except:
        return ""

def get_price(soup):
    try:
        price = soup.find("div", class_="Nx9bqj CxhGGd").text.strip()
        return price
    except:
        return ""

def get_rating(soup):
    try:
        rating = soup.find("div", class_="XQDdHH").text.strip()
        return rating
    except:
        return ""

def get_review_count(soup):
    try:
        review_count = soup.find("span", class_="Wphh3N").text.strip()
        return review_count
    except:
        return ""

def get_availability(soup):
    try:
        availability = soup.find("div", class_="ZgI5Je").text.strip()
        return availability
    except:
        return "Not Available"


# ---- Main Script ----
if __name__ == '__main__':

    HEADERS = ({
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
        'Accept-Language': 'en-US, en;q=0.5'
    })

    # Flipkart search results URL
    URL = "https://www.flipkart.com/search?q=iphone+16+pro+max"

    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS)
    soup = BeautifulSoup(webpage.content, "html.parser")

    # Fetch product links (anchor tags)
    links = soup.find_all("a", attrs={'class': 'CGtC98'})

    # Store product links
    links_list = []
    for link in links:
        href = link.get('href')
        if href:
            links_list.append("https://www.flipkart.com" + href)

    d = {"title": [], "price": [], "rating": [], "reviews": [], "availability": []}

    # Loop for extracting product details
    for link in links_list:
        new_webpage = requests.get(link, headers=HEADERS)
        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['reviews'].append(get_review_count(new_soup))
        d['availability'].append(get_availability(new_soup))

    # Save to CSV
    flipkart_df = pd.DataFrame.from_dict(d)
    flipkart_df['title'].replace('', np.nan, inplace=True)
    flipkart_df = flipkart_df.dropna(subset=['title'])
    flipkart_df.to_csv("flipkart_data.csv", header=True, index=False)

    print(" Data saved to flipkart_data.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("flipkart_data.csv")

# Clean the price column (remove  and commas, convert to float)
df['price_clean'] = df['price'].astype(str).str.replace('','').str.replace(',','').astype(float)

# Clean rating column (convert to float, handle missing values)
df['rating_clean'] = df['rating'].astype(str).str.split(' ').str[0]
df['rating_clean'] = pd.to_numeric(df['rating_clean'], errors='coerce')  # convert to float, NaN if invalid

# --- 1. Price Distribution ---
plt.figure(figsize=(10,6))
plt.hist(df['price_clean'].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.title('Price Distribution of Products')
plt.xlabel('Price (INR)')
plt.ylabel('Number of Products')
plt.grid(axis='y', alpha=0.75)
plt.show()

# --- 2. Rating vs Price Scatter Plot ---
plt.figure(figsize=(10,6))
plt.scatter(df['price_clean'], df['rating_clean'], color='orange')
plt.title('Product Rating vs Price')
plt.xlabel('Price (INR)')
plt.ylabel('Rating (out of 5)')
plt.grid(True)
plt.show()

# --- 3. Top 10 Most Expensive Products ---
top10 = df.sort_values(by='price_clean', ascending=False).head(10)
plt.figure(figsize=(12,6))
plt.barh(top10['title'], top10['price_clean'], color='green')
plt.title('Top 10 Most Expensive Products')
plt.xlabel('Price (INR)')
plt.gca().invert_yaxis()  # Highest price on top
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("flipkart_data.csv")

# Clean the price column
df['price_clean'] = df['price'].astype(str).str.replace('','').str.replace(',','').astype(float)

# Clean the rating column
df['rating_clean'] = df['rating'].astype(str).str.split(' ').str[0]
df['rating_clean'] = pd.to_numeric(df['rating_clean'], errors='coerce')

# Sort data by price for better line chart visualization
df_sorted = df.sort_values(by='price_clean')

# --- 1. Price Trend Line Chart ---
plt.figure(figsize=(12,6))
plt.plot(df_sorted['title'], df_sorted['price_clean'], marker='o', linestyle='-', color='blue')
plt.xticks(rotation=90)
plt.title('Product Price Trend')
plt.xlabel('Product')
plt.ylabel('Price (INR)')
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 2. Rating Trend Line Chart ---
plt.figure(figsize=(12,6))
plt.plot(df_sorted['title'], df_sorted['rating_clean'], marker='o', linestyle='-', color='orange')
plt.xticks(rotation=90)
plt.title('Product Rating Trend')
plt.xlabel('Product')
plt.ylabel('Rating (out of 5)')
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 3. Price vs Rating Line Chart ---
plt.figure(figsize=(12,6))
plt.plot(df_sorted['price_clean'], df_sorted['rating_clean'], marker='o', linestyle='-', color='green')
plt.title('Rating vs Price')
plt.xlabel('Price (INR)')
plt.ylabel('Rating (out of 5)')
plt.grid(True)
plt.show()
```
